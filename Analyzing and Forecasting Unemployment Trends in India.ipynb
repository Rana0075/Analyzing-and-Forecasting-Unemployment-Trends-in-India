```python
df2=df1.copy()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]
```

```python
df2 = remove_outliers_iqr(df2, 'Estimated Employed')
df2 = remove_outliers_iqr(df2, 'Estimated Labour Participation Rate (%)')
```

```python
sns.jointplot(y=df2['Estimated Employed'],x=df2['Estimated Labour Participation Rate (%)'])
```

```python
df2.info()
```

```python
# Computing average unemployment by Area
gap= df2.groupby(['Region', 'Area'])['Estimated Unemployment Rate (%)'].mean().unstack()

# Ruralâ€“Urban Gap
df2['Rural-Urban Gap'] = gap['Rural'] - gap['Urban']
```
