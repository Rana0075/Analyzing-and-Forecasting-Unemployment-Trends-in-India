```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]
```

```python
df2 = remove_outliers_iqr(df2, 'Estimated Employed')
df2 = remove_outliers_iqr(df2, 'Estimated Labour Participation Rate (%)')
```

```python
sns.jointplot(y=df2['Estimated Employed'],x=df2['Estimated Labour Participation Rate (%)'])
```

```python
df2.info()
```

```python
#pivot table
rural_urban_rates = df2.pivot_table(
    index=['Region', 'Year', 'Month'],
    columns='Area',
    values='Estimated Unemployment Rate (%)'
).reset_index()

#Rural Urban Gap
rural_urban_rates['Rural-Urban Gap'] = rural_urban_rates['Rural'] - rural_urban_rates['Urban']

#Merge 
df2 = df2.merge(
    rural_urban_rates[['Region', 'Year', 'Month', 'Rural-Urban Gap']],
    on=['Region', 'Year', 'Month'],
    how='left'
)
```

```python
df2.head(15)
```

Positive gap = Rural unemployment rate is higher than Urban.
Negative gap = Urban unemployment rate is higher than Rural.
